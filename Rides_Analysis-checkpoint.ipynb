### Loading the dataset and getting started

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
csv_path = ("ridesupd.csv")
df = pd.read_csv(csv_path, parse_dates = ['Started at','Finished at'])
df.head()
```

### Getting to know about the data

```python
df.info()
```

### Looking for Unfinised rides

```python
df[df.Status != 'finished']
```

### Looking for Rides with duration less than a minute.

Rides less than a minute are ignored since those rides have been used for testing the cycle's condition.

```python
df['Duration'].loc[df.Duration < '00:01:00'].count() 
```

```python
df.loc[df.Duration < '00:01:00']
```

### Dropping rides with duration less than a minute

```python
new_df = df.drop(df[df.Duration < '00:01:00'].index)
new_df.info()
```

### Checking for rides with Ride Distance as 0.0

```python
new_df.loc[new_df['Ride Distance'] == 0.0]
```

### Dropping values with Ride Distances that are 0

```python
df1 = new_df.drop(new_df[new_df['Ride Distance'] == 0].index)
```

### Looking for highest Ride Distance

```python
df1.sort_values(by='Ride Distance',ascending = False)
```

### Looking for missing Vehicle ID


Ignoring vehicles with no Vehicle ID

```python
df1.loc[df1['Vehicle ID'].isnull() == True]
```

### Taking only finite values of Vehicle IDs

```python
df1 = df1[np.isfinite(df1['Vehicle ID'])]
```

```python
df1.info()
```

### Creating new columns for month and year

```python
df1['Month'] = pd.DatetimeIndex(df1['Started at']).month
df1['Year'] = pd.DatetimeIndex(df1['Started at']).year
df1
```

### Fare each month

No data for the months 4,5,6,7,8 and 9 hence these months do not have any users shown. More specifically speaking, the Vehicle IDs have null values for these months and since such Vehicle IDs have been ignored these months do not show up in our report.

```python
df1.groupby(['Month','Year']).agg({'Fare':np.sum})
```

### Finding top 3 users for every month


No data for the months 4,5,6,7,8 and 9 hence these months do not have any users shown. More specifically speaking, the Vehicle IDs have null values for these months and since such Vehicle IDs have been ignored these months do not show up in our report.

```python
group1 = df1.groupby(['Year','Month','Member ID']).agg({'ID': 'count', 'Ride Distance': np.sum,'Fare': np.sum})
group1.sort_values('ID',ascending = False).sort_index(level = [0],ascending = [True]).sort_index(level = [1],ascending = [True]).reset_index(level=[1,2]).groupby(['Year','Month']).head(3)

```

```python
group1 = df1.groupby(['Year','Month','Member ID']).agg({'ID': 'count', 'Ride Distance': np.sum,'Fare': np.sum})
group1.sort_values('ID',ascending = False).sort_index(level = [0,1],ascending = [True,True]).reset_index(level = [1,2]).groupby(['Year','Month']).head(3)
```

### Finding top 5 most used cycles

i.e The number of times a particular cycle has been used.

```python
group2 = df1.groupby(['Year','Month','Vehicle ID']).agg({'ID': 'count', 'Ride Distance': np.sum,'Fare': np.sum})
group2.sort_values('ID',ascending = False).sort_index(level = [0,1],ascending = [True,True]).reset_index(level = [1,2]).groupby(['Year','Month']).head(5)
```

### Finding top 5 least used cycles.

```python
group2 = df1.groupby(['Year','Month','Vehicle ID']).agg({'ID': 'count', 'Ride Distance': np.sum,'Fare': np.sum})
group2.sort_values('ID',ascending = True).sort_index(level = [0,1],ascending = [True,True]).reset_index(level = [1,2]).groupby(['Year','Month']).head(5)
```

## Data Visualization

### Looking for Correlation between different fields

```python
corr = df1.corr()

sb.heatmap(
    data=corr,
    annot=True,
    linewidths=0.5,
    linecolor = 'pink',
    cmap='BuPu'
)

fig = plt.gcf()
fig.set_size_inches(30, 10)
plt.show()


```

### Number of rides taken per month in each year

Each ride has a uniqe ID which is identified by "ID" in this dataset.

```python
df1.groupby(['Month','Year']).agg({'ID': 'count'}).plot.bar(color = 'c')
```

### Total Fare generated per month in each year

```python
df1.groupby(['Month','Year']).agg({'Fare': np.sum}).plot.bar(color = 'orange')
```

### Member Type distribution in the population

```python
fig1, ax1 = plt.subplots()
ax1.pie(df1.groupby('Member Type').agg({'ID':'count'}), labels=['Casual User','Subscribed User'], autopct='%1.1f%%',
        shadow=True, startangle=90 )
ax1.axis('equal') 
plt.show()
```
